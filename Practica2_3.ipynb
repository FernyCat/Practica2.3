```python
import numpy as np
import pandas as pd
```

```python
#Archivo
data = pd.read_csv('/content/drive/MyDrive/Wine.csv', header=None)
data.head()
```

## Eliminamos la primera columna

```python
# 1- Eliminar primera columna
data = data.drop(0,axis=1)
#Agregar nombres a las columnas para facilitar su uso
columns = ['X1','X2','X3','X4','X5','X6','X7','X8','X9','X10','X11','X12','X13']
data.columns = columns
```

### Normalizacion

```python
# Metodo min max
new_data = data
for col in new_data.columns:
  new_data[col] = (data[col] - data[col].min()) / (data[col].max() - data[col].min())

new_data.head()
```

## Matriz de distancias

```python
import math

n_points = len(new_data)
distance_matrix = np.zeros((n_points, n_points))

for i in range(n_points):
    for j in range(i + 1, n_points):
        point1 = new_data.iloc[i]
        point2 = new_data.iloc[j]

        sum = 0

        for col in new_data.columns:
            sum += (point1[col] - point2[col]) ** 2

        euclidean_distance = math.sqrt(sum)
        distance_matrix[i][j] = euclidean_distance
        distance_matrix[j][i] = euclidean_distance

distance_matrix
```

```python
#distancia al cuadrado
distancia_2 = distance_matrix ** 2
distancia_2
```

## Matriz centalizada

```python
#doble centrado
n = len(distance_matrix)
j = np.ones(n)
I = np.eye(n)
C = I - ((j)/n)
C
```

```python
B = (-1/2)* np.dot(np.dot(C,distancia_2),C)
B
```
